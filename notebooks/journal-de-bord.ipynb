### 🔧 Étapes / Actions réalisées

### 📝 Observations / Difficultés
- Vigilance fuite de cible: `casual_riders` et `registered_riders` composent directement `total_rentals`.
- Écart fuite vs pipeline faible sur ce volume de données (~17k), mais règle anti-fuite maintenue par défaut.
- Choix justifié: `hist_gradient_boosting` retenu car meilleur compromis précision/coût (RMSE CV 75.58, RMSE hold-out 70.44).

### 🎯 Choix justifié (C4)
J'ai retenu `hist_gradient_boosting` car il offre le meilleur compromis pour le cas Mistral: précision supérieure (`RMSE CV 75.58`, `R² 0.7996`, `RMSE hold-out 70.44`) tout en gardant un coût d'entraînement raisonnable et une taille modèle faible (<1 Mo). Ce choix est cohérent avec la saisonnalité marquée et les relations non linéaires observées en EDA.
